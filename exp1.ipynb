```python
import pandas as pd
import numpy as np
Series_A=pd.Series([10,20,30,40,50])
Series_B=pd.Series([40,50,60,70,80])
union=pd.Series(np.union1d(Series_A,Series_B))
intersect=pd.Series(np.intersect1d(Series_A,Series_B))
notcommonseries=union[-union.isin(intersect)]
print(notcommonseries)
print("largest number in series_A:",Series_A.max())
print("smallest number in series_A:",Series_A.min())
print("the sum of seies_B is:",Series_B.sum())
print("the average of series_A:",Series_A.mean())
print("the median of series_B:",Series_B.median())
```
