```python
import pandas as pd
import numpy as np
data={
    'first':['aryan','rohan','riya','yash','siddini'],
    'lastname':['singh','agarwal','shah','batia','khanna'],
    'type':['full time employee','intern','full-time employee','part time employee','full-time employee'],
    'department':['adminsistration','technical','administration','technical','technical'],
    'yoe':[2,3,5,7,6],
    'salary':[20000,5000,10000,20000,10000]
    }
df=pd.DataFrame(data)
print(df)
table=pd.pivot_table(df,index=['type','department'],
  values='salary',aggfunc=['mean'])
print(table)
table1=pd.pivot_table(df,index=['type'],
  values='salary',aggfunc=['mean','sum'])
print(table1)
table2=pd.pivot_table(df,index=['type'],
  values='salary',aggfunc=['count'])
print(table2)
table3=pd.pivot_table(df,index=['type'],
  values='salary',aggfunc=['std'])
print(table3)


```
